```python
import matplotlib.pyplot as plt
import numpy as np
# ⽣成⼀些数据
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.random.rand(100)
fig, axs = plt.subplots(3, 1, figsize=(8, 12))
# 绘制折线图
axs[0].plot(x, y1, label='sin(x)', color='blue')
axs[0].set_title('1023040905')
axs[0].legend()
# 绘制柱状图
bars = ['A', 'B', 'C', 'D']
heights = [3, 7, 5, 8]
axs[1].bar(bars, heights, color='green')
axs[1].set_title('1023040905')
# 绘制散点图
axs[2].scatter(x, y3, color='red')
axs[2].set_title('1023040905')
plt.tight_layout()
plt.show()
```
